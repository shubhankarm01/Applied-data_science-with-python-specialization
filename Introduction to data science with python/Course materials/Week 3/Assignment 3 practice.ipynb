```python
import pandas as pd
import numpy as np
```

```python
Energy = pd.read_excel('Energy Indicators.xls')
Energy = Energy[(Energy.index > 16) & (Energy.index < 244)]
Energy.head()
```

```python
Energy = Energy.reset_index()
Energy = Energy.drop(['index','Unnamed: 0','Unnamed: 1'], axis = 1)
#Energy.columns.unique()
#Energy = Energy[['Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4', 'Unnamed: 5']]
Energy.head()
```

```python
Energy = Energy.rename(columns = {'Unnamed: 2':'Country', 'Unnamed: 3':'Energy Supply', 'Unnamed: 4':'Energy Supply per Capita', 'Unnamed: 5':'% Renewable'})
Energy.head()
```

```python
Energy = Energy.set_index('Energy Supply')
#Energy.head()
```

```python
def type1(item):
    if item == '...':
        return 0
    return 1
    
for i,j in Energy.groupby(type1):
    j = j.reset_index()
    if i == 1:
        j['Energy Supply'] = j['Energy Supply']*10**6
        g1 = j.copy()
    else:
        j['Energy Supply'] = np.NaN
        g2 = j.copy()
 
#g1.head()
#g2.head()
#pd.merge(g1,g2, how = 'outer', left_on = ['Country', 'Energy Supply', 'Energy Supply per Capita', '% Renewable'], right_on = ['Country', 'Energy Supply', 'Energy Supply per Capita', '% Renewable']).head()
Energy = pd.concat([g1,g2]).reset_index()
Energy = Energy.drop('index', axis = 1)
#Energy.head()
```

```python
def Namemod(item):
    import re
    if item[-1] == ')':
        return(re.split('\s\(', item)[0])
    if item[-1].isnumeric():
        return re.split('\d', item)[0]
    else:
        return item
    
Energy['Country'] = Energy['Country'].apply(Namemod)
```

```python
Energy['Country'] = Energy['Country'].replace(['Republic of Korea', 'United States of America', 'United Kingdom of Great Britain and Northern Ireland', 'China, Hong Kong Special Administrative Region'],['South Korea', 'United States', 'United Kingdom', 'Hong Kong'])
Energy.head()
```

```python
GDP = pd.read_csv('world_bank.csv', skiprows = 4)
#GDP.head()
```

```python
GDP['Country Name'] = GDP['Country Name'].replace(['Korea, Rep.', 'Iran, Islamic Rep.', 'Hong Kong SAR, China'],['South Korea', 'Iran', 'Hong Kong'])
```

```python
GDP = GDP[['Country Name', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015']]
```

```python
df = pd.merge(pd.merge(GDP, Energy, how = 'inner', left_on = 'Country Name', right_on = 'Country'), ScimEn, how = 'inner', left_on = 'Country', right_on = 'Country')
df
```

# Question 11

```python
dftc = []

ContinentDict  = {'China':'Asia', 
                  'United States':'North America', 
                  'Japan':'Asia', 
                  'United Kingdom':'Europe', 
                  'Russian Federation':'Europe', 
                  'Canada':'North America', 
                  'Germany':'Europe', 
                  'India':'Asia',
                  'France':'Europe', 
                  'South Korea':'Asia', 
                  'Italy':'Europe', 
                  'Spain':'Europe', 
                  'Iran':'Asia',
                  'Australia':'Australia', 
                  'Brazil':'South America'}

def groupbycont(item):
    if item in ContinentDict:
        return ContinentDict[item]

for i,j in df.groupby(groupbycont):
    
    j['Continent'] = i
    j = j.set_index('Continent')
    
    j['size'] = len(j)
    j['sum'] = np.sum(j['Population'])
    j['mean'] = np.mean(j['Population'])
    j['std'] = np.std(j['Population'])
    
    dftc.append(j)
    
df2 = pd.concat(dftc)
df2 = df2[['size', 'sum', 'mean', 'std']]
```

```python
dftc1 = []

ContinentDict  = {'China':'Asia', 
                  'United States':'North America', 
                  'Japan':'Asia', 
                  'United Kingdom':'Europe', 
                  'Russian Federation':'Europe', 
                  'Canada':'North America', 
                  'Germany':'Europe', 
                  'India':'Asia',
                  'France':'Europe', 
                  'South Korea':'Asia', 
                  'Italy':'Europe', 
                  'Spain':'Europe', 
                  'Iran':'Asia',
                  'Australia':'Australia', 
                  'Brazil':'South America'}

def groupbycont(item):
    if item in ContinentDict:
        return ContinentDict[item]

for i,j in df.groupby(pd.cut(df['% Renewable'],5)):
    for m,n in j.groupby(groupbycont):
        n['Continent'] = m
        n['% Renewable bin'] = i
        n['countries count'] = len(n)
        
        dftc1.append(n)    
    
df3 = pd.concat(dftc1)
df3 = df3.set_index(['Continent', '% Renewable bin'])
df3 = df3['countries count']
```
