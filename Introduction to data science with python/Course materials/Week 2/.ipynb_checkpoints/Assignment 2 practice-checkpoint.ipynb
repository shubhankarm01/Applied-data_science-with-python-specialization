```python
import pandas as pd

df = pd.read_csv('olympics.csv', index_col=0, skiprows=1)

for col in df.columns:
    if col[:2]=='01':
        df.rename(columns={col:'Gold'+col[4:]}, inplace=True)
    if col[:2]=='02':
        df.rename(columns={col:'Silver'+col[4:]}, inplace=True)
    if col[:2]=='03':
        df.rename(columns={col:'Bronze'+col[4:]}, inplace=True)
    if col[:1]=='№':
        df.rename(columns={col:'#'+col[1:]}, inplace=True)

df.head()
```

```python
names_ids = df.index.str.split('\s\(') # split the index by '('

df.index = names_ids.str[0] # the [0] element is the country name (new index) 
df['ID'] = names_ids.str[1].str[:3] # the [1] element is the abbreviation or ID (take first 3 characters from that)

df = df.drop('Totals')
df.head()
```

```python
def answer_one():
    df2 = df.copy()
    df2['Country'] = df.index
    df2 = df2.set_index('Gold')
    df2 = df2.sort_index()
    return df2.iloc[len(df2)-1]['Country']

answer_one()
```

```python
def answer_two():
    df3 = df.copy()
    df3['difference'] = abs(df3['Gold']-df3['Gold.1'])
    df3['Country'] = df3.index
    df3 = df3.set_index('difference')
    df3 = df3.sort_index()
    return df3.iloc[len(df3)-1]['Country']

answer_two()    
```

```python
def answer_three():
    df4 = df[(df['Gold'] > 0) & (df['Gold.1'] > 0)]
    df4['drel'] = abs(df4['Gold'] - df4['Gold.1'])/df4['Gold.2']
    df4['Country'] = df4.index
    df4 = df4.set_index('drel')
    df4 = df4.sort_index()
    return df4.iloc[len(df4)-1]['Country']

answer_three()
```

```python
def answer_four():
    df5 = df.copy()
    df5['Points'] = (df5['Gold.2']*3)+(df5['Silver.2']*2)+(df5['Bronze.2']*1)
    return df5['Points']

answer_four()
```
