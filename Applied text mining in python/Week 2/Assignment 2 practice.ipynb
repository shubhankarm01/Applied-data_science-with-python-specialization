```python
import nltk
import pandas as pd
import numpy as np

# If you would like to work with the raw text you can use 'moby_raw'
with open('moby.txt', 'r') as f:
    moby_raw = f.read()
    
# If you would like to work with the novel in nltk.Text format you can use 'text1'
moby_tokens = nltk.word_tokenize(moby_raw)
text1 = nltk.Text(moby_tokens)
```

```python
def answer_three():
    
#     from nltk.stem import WordNetLemmatizer
#     from nltk.stem import WordNetLemmatizer
#     lemmatizer = WordNetLemmatizer()
    
#     lemmatize = [lemmatizer.lemmatize(w) for w in text1]
    
    unique_token = set(text1)
    text1_freq_dist = nltk.FreqDist(text1)
    unique_token_freq = [text1_freq_dist[w] for w in unique_token]
    unique_token_dist = dict(zip(unique_token, unique_token_freq))
    
    i = sorted(unique_token_dist.values(), reverse = True)[:20]
    
    list1 = []
    
    for keys, values in unique_token_dist.items():
        if values in i:
            list1.append((keys, values))
    
    return(sorted(list1, key= lambda x: x[1], reverse = True))
#     return unique_token_freq
    
answer_three()
```

```python
def answer_five():
    
    words = list(w for w in text1)
    length = list(len(w) for w in words)

    i = dict(zip(words, length))
    
    j = sorted(i.values(), reverse = True)[0]
    
    for keys, values in i.items():
        if values == j:
            return((keys, j))
    
answer_five()
```

```python
def answer_six():
    
    distribution = nltk.FreqDist(text1)
    
    i = [w for w in distribution.keys() if distribution[w] > 2000 and w.isalpha() == True]
    j = [distribution[m] for m in i]
    
    return list(sorted(zip(i, j), key = lambda x : x[1], reverse = True))
    
answer_six()
```

```python
def answer_seven():
    
    sentences = [len(nltk.word_tokenize(i)) for i in moby_raw.splitlines()]
    
    return np.average(sentences)

answer_seven()
```

```python
def answer_eight():
    
    POS = nltk.pos_tag(text1)
    
    token, POS1 = list(zip(* POS))
    
    POS2 = set(POS1)
    
    return sorted(list(zip(POS2, list(POS1.count(x) for x in POS2))), key = lambda m:m[1], reverse = True)[:5]

answer_eight()
```

```python
from nltk.corpus import words

correct_spellings = words.words()
```

```python
correct_spellings
```

```python
def answer_nine(entries=['cormulent', 'incendenece', 'validrate']):
    
    from nltk.metrics.distance import jaccard_distance
    from nltk.util import ngrams
    
    spellings = pd.Series(correct_spellings)
    
    output = []
    
    for entry in entries:
        match_list = spellings[spellings.str.startswith(entry[0])]
        output.append(min(list((jaccard_distance(set(ngrams(entry, 3)), set(ngrams(word, 3))), word) for word in match_list))[1])
    
    return output
    
answer_nine()
```
