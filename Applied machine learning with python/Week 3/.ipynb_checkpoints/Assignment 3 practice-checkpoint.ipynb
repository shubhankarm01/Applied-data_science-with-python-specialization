```python
import pandas as pd

df = pd.read_csv('fraud_data.csv')
df.head()
```

```python
import numpy as np

list1 = np.bincount(df['Class'])
list1[1]/(list1[0]+list1[1])
```

```python
from sklearn.model_selection import train_test_split

x = df.iloc[:, :-1]
y = df.iloc[:, -1]

x_train, x_test, y_train, y_test = train_test_split(x, y, random_state = 0)
```

```python
from sklearn.dummy import DummyClassifier
from sklearn.metrics import accuracy_score, recall_score, precision_score

dum_cls = DummyClassifier('most_frequent').fit(x_train, y_train)
dum_cls_predict = dum_cls.predict(x_test)

accuracy_score(y_test, dum_cls_predict)
recall_score(y_test, dum_cls_predict)
precision_score(y_test, dum_cls_predict)
```

```python
from sklearn.svm import SVC

svm1 = SVC().fit(x_train, y_train)
svm1_predict = svm1.predict(x_test)

accuracy_score(y_test, svm1_predict)
recall_score(y_test, svm1_predict)
precision_score(y_test, svm1_predict)
```

```python
svm2 = SVC(C = 1e9, gamma = 1e-7).fit(x_train, y_train)
svm2_predict = svm2.predict(x_test)

accuracy_score(y_test, svm2_predict)
recall_score(y_test, svm2_predict)
precision_score(y_test, svm2_predict)
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression().fit(x_train, y_train)
lr_prob = lr.predict_proba(x_test)
```

```python
lr_prob[:, 1]
```

```python
from sklearn.metrics import precision_recall_curve

precision, recall,  threshold = precision_recall_curve(y_test, lr_prob[:,1])
```

```python
import matplotlib.pyplot as plt
%matplotlib notebook

plt.figure()
plt.plot(precision, recall)
```

```python
from sklearn.metrics import roc_curve

fpr, tpr, _ = roc_curve(y_test, lr_prob[:, 1])

plt.figure()
plt.plot(fpr, tpr)
```

```python
from sklearn.model_selection import GridSearchCV

param_values = {'penalty': ['l1', 'l2'], 'C':[0.01, 0.1, 1, 10, 100]}

grid = GridSearchCV(LogisticRegression(), cv = 3, param_grid = param_values, scoring = 'recall')
grid.fit(x_train, y_train).cv_results_['mean_test_score'].reshape(5,2)
```
