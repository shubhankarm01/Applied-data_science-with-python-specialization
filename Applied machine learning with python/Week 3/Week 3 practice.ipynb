```python
from sklearn.datasets import load_digits
```

```python
ds = load_digits()
```

```python
x, y = ds.data, ds.target
```

```python
y[y != 1] = 0
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x,y)
```

```python
from sklearn.svm import SVC

svm = SVC(kernel = 'rbf', C=1).fit(X_train, y_train)
svm.score(X_test, y_test)
```

```python
from sklearn.dummy import DummyClassifier

dc = DummyClassifier(strategy = 'most_frequent').fit(X_train, y_train)
y_dp = dc.predict(X_test)
```

```python
dc.score(X_test, y_test)
```

```python
svm = SVC(kernel = 'linear', C = 1).fit(X_train, y_train)
svm.score(X_test, y_test)
```

```python
svmp = svm.predict(X_test)
svmp
```

```python
from sklearn.metrics import confusion_matrix

confusion = confusion_matrix(y_test, y_dp)
```

```python
confusion
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, svmp))
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression().fit(X_train, y_train)
```

```python
lr.score(X_test, y_test)
```

```python
y_score_lr = lr.fit(X_train, y_train).decision_function(X_test)

y_score_list = list(zip(y_test[0:20], y_score_lr[0:20]))

y_score_list
```

```python
from sklearn.metrics import precision_recall_curve
import numpy as np

precision, recall, thresholds =precision_recall_curve(y_test, y_score_lr)

closest_zero = np.argmin(np.abs(thresholds))
closest_zero_p = precision[closest_zero]
closest_zero_r = recall[closest_zero]
closest_zero_p
```

```python
import matplotlib.pyplot as plt
%matplotlib notebook

plt.figure()
plt.plot(precision, recall)
```
