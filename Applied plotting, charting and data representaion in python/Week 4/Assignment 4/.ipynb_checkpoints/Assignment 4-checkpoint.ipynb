# First dataset for total energy consumption

```python
import pandas as pd
```

```python
ECbySector = pd.read_csv('Energy consumption by sector.csv')
```

```python
ECbySector.head()
```

```python
ECbySector = ECbySector[['YYYYMM', 'Value', 'Description']]
ECbySector = ECbySector[(ECbySector['YYYYMM'] > 195000)]
ECbySector = ECbySector[(ECbySector['YYYYMM'] < 202000)]
ECbySector.head()
```

## Cleaning data

```python
import re
ECbySector['Decade'] = ECbySector['YYYYMM'].apply(lambda item : str(item)[0:3]+'0')
ECbySector['Month'] = ECbySector['YYYYMM'].apply(lambda item : str(item)[4:6])
ECbySector['Energy type'] = ECbySector['Description'].apply(lambda item : re.split(' ', item)[0])
ECbySector['Sector'] = ECbySector['Description'].apply(lambda item : re.split(' ', item)[-2])
ECbySector = ECbySector[ECbySector['Energy type'] == 'Total']
ECbySector = ECbySector[ECbySector['Month'] < '13']
ECbySector.head()
```

```python
ECbySector = ECbySector.pivot_table(index = 'Decade', columns = 'Sector', values = 'Value', aggfunc = sum)
```

```python
ECbySector
```

```python
import matplotlib.pyplot as plt

%matplotlib notebook
```

```python
ECbySector.plot.bar(stacked = True, width = 0.8)
```

# 2nd and 3rd dataset for Renewable energy consumption

```python
RECbyCommRes = pd.read_csv('RE consumption by residential and commercial sector.csv')
RECbyIndTrans = pd.read_csv('RE consumption by industrial and transport sector.csv')
```

## Cleaning data

```python
RECbyIndTrans['Decade'] = RECbyIndTrans['YYYYMM'].apply(lambda item : str(item)[0:3]+'0')
RECbyIndTrans['Month'] = RECbyIndTrans['YYYYMM'].apply(lambda item : str(item)[4:6])
RECbyIndTrans['Energy type'] = RECbyIndTrans['Description'].apply(lambda item : re.split(' ', item)[0])
RECbyIndTrans['Sector'] = RECbyIndTrans['Description'].apply(lambda item : re.split(' ', item)[-2])
RECbyIndTrans = RECbyIndTrans[RECbyIndTrans['Energy type'] != 'Total']
RECbyIndTrans = RECbyIndTrans[RECbyIndTrans['Month'] < '13']
```

```python
RECbyIndTrans = RECbyIndTrans[RECbyIndTrans['YYYYMM'] > 195000]
RECbyIndTrans = RECbyIndTrans[RECbyIndTrans['YYYYMM'] < 202000]
RECbyIndTrans = RECbyIndTrans[RECbyIndTrans['Value'] != 'Not Available']
RECbyIndTrans = RECbyIndTrans[RECbyIndTrans['Value'] != 'No Data Reported']
RECbyIndTrans['Value'] = RECbyIndTrans['Value'].astype(float)
```

```python
RECbyIndTrans.head()
```

```python
RECbyIndTrans['Sector'] = RECbyIndTrans['Sector'].apply(lambda item : 'Transportation' if item == 'Fuels' else item)
```

```python
RECbyIndTrans = RECbyIndTrans.pivot_table(index = 'Decade', columns = 'Sector', values = 'Value', aggfunc = sum)
```

```python
RECbyIndTrans
```

```python
RECbyCommRes = RECbyCommRes[['YYYYMM', 'Value', 'Description']]
RECbyCommRes = RECbyCommRes[RECbyCommRes['YYYYMM'] > 195000]
RECbyCommRes = RECbyCommRes[RECbyCommRes['YYYYMM'] < 202000]
RECbyCommRes.head()
```

# cleaning data

```python
import re
RECbyCommRes['Decade'] = RECbyCommRes['YYYYMM'].apply(lambda item : str(item)[0:3]+'0')
RECbyCommRes['Month'] = RECbyCommRes['YYYYMM'].apply(lambda item : str(item)[4:6])
RECbyCommRes['Energy type'] = RECbyCommRes['Description'].apply(lambda item : re.split(' ', item)[0])
RECbyCommRes['Sector'] = RECbyCommRes['Description'].apply(lambda item : re.split(' ', item)[-2])
RECbyCommRes = RECbyCommRes[RECbyCommRes['Month'] < '13']
RECbyCommRes = RECbyCommRes[RECbyCommRes['Energy type'] == 'Total']
RECbyCommRes['Value'] = RECbyCommRes['Value'].astype(float)
RECbyCommRes.head()
```

```python
# RECbyCommRes['Description'].unique()
```

```python
import numpy as np
RECbyCommRes = RECbyCommRes.pivot_table(index = 'Decade', columns = 'Sector', values = 'Value', aggfunc = sum)
```

```python
RECbyCommRes
```

## Merging datasets

```python
RECbySect = pd.merge(RECbyCommRes, RECbyIndTrans, how = 'outer', left_on = 'Decade', right_on = 'Decade')
```

```python
RECbySect
```

```python
plt.figure();

RECbySect.plot.bar(stacked = True);
```

# 4th, 5th, 6th and 7th dataset for CO2 emission

```python
CO2emmbyComm = pd.read_csv('Commercial sector CO2 emission.csv')
CO2emmbyComm['Sector'] = 'Commercial'
CO2emmbyInd = pd.read_csv('Industrial sector CO2 emission.csv')
CO2emmbyInd['Sector'] = 'Industrial'
CO2emmbyRes = pd.read_csv('Residential sector CO2 emission.csv')
CO2emmbyRes['Sector'] = 'Residential'
CO2emmbyTrans = pd.read_csv('Transport sector CO2 emission.csv')
CO2emmbyTrans['Sector'] = 'Transporation'
```

```python
CO2emmbySect = pd.concat([CO2emmbyComm, CO2emmbyInd, CO2emmbyRes, CO2emmbyTrans])
```

```python
CO2emmbySect.head()
```

```python
CO2emmbySect = CO2emmbySect[['YYYYMM', 'Value', 'Description', 'Sector']]
CO2emmbySect = CO2emmbySect[CO2emmbySect['YYYYMM'] > 195000]
CO2emmbySect = CO2emmbySect[CO2emmbySect['YYYYMM'] < 202000]
CO2emmbySect.head()
```

```python
import re
CO2emmbySect['Decade'] = CO2emmbySect['YYYYMM'].apply(lambda item : str(item)[0:3]+'0')
CO2emmbySect['Month'] = CO2emmbySect['YYYYMM'].apply(lambda item : str(item)[4:6])
CO2emmbySect['Energy type'] = CO2emmbySect['Description'].apply(lambda item : re.split(' ', item)[0])
# CO2emmbySect['Sector'] = CO2emmbySect['Description'].apply(lambda item : re.split(' ', item)[-2])
CO2emmbySect = CO2emmbySect[CO2emmbySect['Month'] < '13']
CO2emmbySect = CO2emmbySect[CO2emmbySect['Energy type'] == 'Total']
CO2emmbySect['Value'] = CO2emmbySect['Value'].astype(float)
CO2emmbySect.head()
```

```python
CO2emmbySect = CO2emmbySect.pivot_table(index = 'Decade', columns = 'Sector', values = 'Value', aggfunc = sum)
```

```python
CO2emmbySect
```

```python
CO2emmbySect.plot.bar(stacked = True)
plt.ylabel('CO2 emission')
```

# Final plotting

```python
import matplotlib.gridspec as gridspec

fig = plt.figure()

gs = gridspec.GridSpec(2,2, hspace = 0.2, wspace = 0.05)

fg = plt.subplot(gs[0,0])
sg = plt.subplot(gs[0,1])
tg = plt.subplot(gs[1,0])
fog = plt.subplot(gs[1,1])
```

```python
ECbySector.plot.bar(stacked = True, ax = fg)
RECbySect.plot.bar(stacked = True, ax = sg)
CO2emmbySect.plot.bar(stacked = True, ax = tg)
```

```python
fg.xaxis.set_visible(False)
sg.xaxis.set_visible(False)
tg.set_xlabel('')
sg.yaxis.set_ticks_position('right')
sg.yaxis.set_label_position('right')
for i in tg.get_xticklabels():
    i.set_rotation(0)
```

```python
fg.set_title('Total energy consumption', fontname = 'Times New Roman')
sg.set_title('Renewable energy consumption', fontname = 'Times New Roman')
tg.set_title('Carbon dioxide emission', fontname = 'Times New Roman')
fog.set_title('Trends', fontname = 'Times New Roman')
```

```python
plt.setp(tg.get_xticklabels(), fontname = 'Times New Roman')
plt.setp(fg.get_yticklabels(), fontname = 'Times New Roman')
plt.setp(sg.get_yticklabels(), fontname = 'Times New Roman')
plt.setp(tg.get_yticklabels(), fontname = 'Times New Roman')
fg.set_ylabel('Trillion Btu', fontname = 'Times New Roman')
# sg.set_ylabel('Trillion Btu', fontname = 'Times New Roman')
tg.set_ylabel('Metric ton', fontname = 'Times New Roman')
# sg.set_ylabel(rotation = 90)
```

```python
fg.legend(fontsize = 4.4)
sg.get_legend().remove()
tg.get_legend().remove()
```

```python
fp = pd.DataFrame()
```

```python
fp['Total Consumption'] = ECbySector.sum(axis = 1)
fp['Renewable energy Consumption'] = RECbySect.sum(axis =1)
fp['CO2 emission'] = CO2emmbySect.sum(axis =1)
```

```python
for i in fp.columns.unique():
    fp[i] = fp[i]/fp[i].max(axis = 0)
```

```python
fp = fp.reset_index()
```

```python
fog.plot(fp['Decade'], fp['Total Consumption'], '--o', label = 'Total consumption')
fog.plot(fp['Decade'], fp['Renewable energy Consumption'] + 0.1, ':X', label = 'Total renewable consumption')
fog.plot(fp['Decade'], fp['CO2 emission'] - 0.1 , '-s', label = 'Total CO2 emission')
fog.legend(loc = 4, fontsize = 6)
fog.yaxis.set_visible(False)
plt.setp(fog.get_xticklabels(), fontname = 'Times New Roman');
```
