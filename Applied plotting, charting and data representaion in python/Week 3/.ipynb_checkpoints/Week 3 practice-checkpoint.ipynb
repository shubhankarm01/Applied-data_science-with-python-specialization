```python
%matplotlib notebook

import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.figure()

plt.subplot(1, 2, 1)

ld = np.array([1,2,3,4,5,6,7,8])

plt.plot(ld, '-o')

plt.subplot(1, 2, 2)

ed = ld**2

plt.plot(ed, '-o')
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, sharex = True)

axs = [ax1, ax2, ax3, ax4]

for n in range(0, len(axs)):
    ss = 10**(n+1)
    s = np.random.normal(loc = 0, scale = 1, size = ss)
    axs[n].hist(s)
    axs[n].set_title('n{}'.format(ss))
```

```python
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(2,2, sharex = True)
axs = [ax1, ax2, ax3, ax4]

for i in range(len(axs)):
    ss = 10**(i+1)
    s = np.random.normal(loc = 0, scale = 1, size =ss)
    axs[i].hist(s, bins = 100)
    axs[i].set_title('n = {}'.format(ss))
```

```python
import matplotlib.gridspec as gridspec

plt.figure()
gspec = gridspec.GridSpec(3,3)

th = plt.subplot(gspec[0, 1:])
sh = plt.subplot(gspec[1: , 0])
lr = plt.subplot(gspec[1:, 1:])
```

```python
y = np.random.normal(loc = 0, scale = 1, size = 1000)
x = np.random.random(size = 1000)

lr.clear()
lr.scatter(x, y)

th.clear()
th.hist(x, bins = 100, normed = True)

sh.clear()
s = sh.hist(y, bins = 100, orientation = 'horizontal', normed = True)
```

```python
sh.invert_xaxis()
```

```python
for ax in [th, lr]:
    ax.set_xlim(0, 1)
for ax in [sh, lr]:
    ax.set_ylim(-5,5)
```

```python
import pandas as pd

ns = np.random.normal(loc = 0, scale = 1, size = 10000)
rs = np.random.random(size = 10000)
gs = np.random.gamma(2, size = 10000)

df = pd.DataFrame({'normal' : ns, 'random' : rs, 'gamma' : gs})
```

```python
df.describe()
```

```python
plt.figure()

_ = plt.boxplot(df['normal'], whis = 'range')
```

```python
plt.clf()

_ = plt.boxplot([df['normal'], df['random'], df['gamma']], whis = 'range')
```

```python
plt.figure()
_ = plt.hist(df['gamma'], bins = 100)
```

```python
plt.figure()

_ = plt.hist(df['random'], bins = 100)
```

```python
import mpl_toolkits.axes_grid1.inset_locator as mpl_il

plt.figure()
plt.boxplot([df['normal'], df['random'], df['gamma']], whis = 'range')

ax2 = mpl_il.inset_axes(plt.gca(), width = '60%', height = '40%', loc = 2)
ax2.hist(df['gamma'], bins = 100)
ax2.margins(x = 0.5)
```

```python
ax2.yaxis.tick_right()
```

```python
plt.figure()
_ = plt.boxplot([df['normal'], df['random'], df['gamma']])
```

```python
import matplotlib.animation as animation

n = 100
x = np.random.randn(n)
```

```python
def update(curr):
    if curr == n: 
        a.event_source.stop()
    plt.cla()
    bins = np.arange(-4, 4, 0.5)
    plt.hist(x[:curr], bins=bins)
    plt.axis([-4,4,0,30])
    plt.annotate('n = {}'.format(curr), [3,27])
```
